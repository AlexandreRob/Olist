```python
import pandas as pd
```

```python
# Load data into dataframes

df_item = pd.read_csv("Jupiter_Notebook/olist_order_items_dataset.csv")
df_reviews = pd.read_csv("Jupiter_Notebook/olist_order_reviews_dataset.csv")
df_orders = pd.read_csv("Jupiter_Notebook/olist_orders_dataset.csv")
df_products = pd.read_csv("Jupiter_Notebook/olist_products_dataset.csv")
df_geolocation = pd.read_csv("Jupiter_Notebook/olist_geolocation_dataset.csv")
df_sellers = pd.read_csv("Jupiter_Notebook/olist_sellers_dataset.csv")
df_order_pay = pd.read_csv("Jupiter_Notebook/olist_order_payments_dataset.csv")
df_customers = pd.read_csv("Jupiter_Notebook/olist_customers_dataset.csv")
df_category = pd.read_csv("Jupiter_Notebook/product_category_name_translation.csv")

# Merge dataframes to create training dataset
df_train = (
    df_orders.merge(df_item, on="order_id", how="left")
    .merge(df_order_pay, on="order_id", how="outer", validate="m:m")
    .merge(df_reviews, on="order_id", how="outer")
    .merge(df_products, on="product_id", how="outer")
    .merge(df_customers, on="customer_id", how="outer")
    .merge(df_sellers, on="seller_id", how="outer")
)

# Print the shape of the training dataset
print(f"The shape of the training dataset is {df_train.shape}.")
```

## Remplacer les city qui ne sont pas avec les bons zip_code

```python
# Create a new DataFrame with a single city for each geolocation_zip_code_prefix
df_zip_city = pd.DataFrame(columns=['geolocation_zip_code_prefix', 'city_and_state'])

for prefix, group in df_location.groupby('geolocation_zip_code_prefix'):
    # Count the number of occurrences of each city
    city_counts = group['geolocation_city_norm'].value_counts()
    
    # Select the most frequent city for this prefix
    most_frequent_city = city_counts.index[0]
    
    # Add the prefix and the most frequent city to a new DataFrame
    df_new = pd.DataFrame({'geolocation_zip_code_prefix': [prefix], 'city_and_state': [most_frequent_city]})
    
    # Concatenate the new DataFrame with df_zip_city
    df_zip_city = pd.concat([df_zip_city, df_new], ignore_index=True)

df_zip_city
```

```python
# Create a dictionary with the prefixes and their corresponding city_and_state combinations
prefixes_dict = {}
for prefix, group in df_location.groupby('geolocation_zip_code_prefix'):
    # Create a set of unique city_and_state combinations for this prefix
    unique_combinations = set(group['geolocation_city_norm'].apply(lambda x: x + '_' + group['geolocation_state'].iloc[0]).unique())
    
    # If there is more than one unique combination, replace the incorrect cities with the most frequent city
    if len(unique_combinations) > 1:
        # Get the most frequent city for this prefix from df_zip_city
        most_frequent_city = df_zip_city.loc[df_zip_city['geolocation_zip_code_prefix'] == prefix, 'city_and_state'].iloc[0]
        
        # Replace the incorrect cities with the most frequent city
        corrected_combinations = []
        for c in unique_combinations:
            parts = c.split('_')
            if len(parts) >= 2 and len(most_frequent_city.split('_')) >= 2 and parts[1] != most_frequent_city.split('_')[1]:
                corrected_combinations.append(most_frequent_city)
            else:
                corrected_combinations.append(c)
        
        # Add the prefix and corrected combinations to the prefixes_dict
        prefixes_dict[prefix] = corrected_combinations
    else:
        # Add the prefix and unique combinations to the prefixes_dict
        prefixes_dict[prefix] = list(unique_combinations)

# Create a new DataFrame with the prefixes and their corresponding city_and_state combinations
prefixes_df = pd.DataFrame(list(prefixes_dict.items()), columns=['geolocation_zip_code_prefix', 'city_and_state_combinations'])

prefixes_df
```

```python
# Créer une colonne temporaire avec la première ville de la liste
prefixes_df['temp_city'] = prefixes_df['city_and_state_combinations'].apply(lambda x: x[0])

# Séparer temp_city en deux colonnes
prefixes_df[['geolocation_city_norm', 'geolocation_state']] = prefixes_df['temp_city'].str.split('_', expand=True)

# Supprimer les colonnes temporaires
prefixes_df.drop(['temp_city', 'city_and_state_combinations'], axis=1, inplace=True)

prefixes_df
```

```python
prefixes_df.insert(0, 'location_id', range(1, 1+len(prefixes_df)))
prefixes_df.set_index('location_id', inplace=True)
prefixes_df
```
